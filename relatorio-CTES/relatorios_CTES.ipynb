# Análise de Dados dos Cursos Técnicos

Este notebook apresenta análises quantitativas dos dados dos Cursos Técnicos da UFRA.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Configuração dos gráficos
plt.style.use('ggplot')
sns.set_palette('husl')
plt.rcParams['figure.figsize'] = (10, 6)
plt.rcParams['font.size'] = 12
```

## Dados de Matrícula por Curso

```python
# Dados simulados para demonstração
cursos_data = {
    'Curso': ['Agropecuária', 'Meio Ambiente', 'Informática', 'Administração', 'Agroindústria'],
    'Matriculas_2022': [45, 38, 42, 35, 28],
    'Matriculas_2023': [52, 44, 48, 41, 33],
    'Matriculas_2024': [58, 47, 51, 43, 36]
}

df_cursos = pd.DataFrame(cursos_data)
print("Dados de Matrícula por Curso:")
print(df_cursos)
```

```python
# Gráfico de evolução das matrículas
fig, ax = plt.subplots(figsize=(12, 7))

x = np.arange(len(df_cursos['Curso']))
width = 0.25

ax.bar(x - width, df_cursos['Matriculas_2022'], width, label='2022', alpha=0.8)
ax.bar(x, df_cursos['Matriculas_2023'], width, label='2023', alpha=0.8)
ax.bar(x + width, df_cursos['Matriculas_2024'], width, label='2024', alpha=0.8)

ax.set_xlabel('Cursos Técnicos')
ax.set_ylabel('Número de Matrículas')
ax.set_title('Evolução das Matrículas nos Cursos Técnicos da UFRA (2022-2024)')
ax.set_xticks(x)
ax.set_xticklabels(df_cursos['Curso'], rotation=45, ha='right')
ax.legend()
ax.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```

## Análise de Desempenho

```python
# Dados de desempenho
desempenho_data = {
    'Curso': ['Agropecuária', 'Meio Ambiente', 'Informática', 'Administração', 'Agroindústria'],
    'Taxa_Aprovacao': [87.5, 89.2, 91.3, 85.7, 88.9],
    'Taxa_Evasao': [8.2, 6.8, 5.1, 9.5, 7.3],
    'Nota_Media': [8.2, 8.5, 8.7, 8.0, 8.3]
}

df_desempenho = pd.DataFrame(desempenho_data)
print("Indicadores de Desempenho:")
print(df_desempenho)
```

```python
# Gráfico de radar para visualizar desempenho
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))

# Gráfico de barras - Taxa de Aprovação
ax1.bar(df_desempenho['Curso'], df_desempenho['Taxa_Aprovacao'], 
        color=['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728', '#9467bd'], alpha=0.8)
ax1.set_title('Taxa de Aprovação por Curso (%)')
ax1.set_ylabel('Taxa de Aprovação (%)')
ax1.set_xticklabels(df_desempenho['Curso'], rotation=45, ha='right')
ax1.grid(True, alpha=0.3)

# Gráfico de dispersão - Relação entre Aprovação e Nota Média
ax2.scatter(df_desempenho['Nota_Media'], df_desempenho['Taxa_Aprovacao'], 
           s=100, alpha=0.7, c=['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728', '#9467bd'])
ax2.set_xlabel('Nota Média')
ax2.set_ylabel('Taxa de Aprovação (%)')
ax2.set_title('Relação entre Nota Média e Taxa de Aprovação')
ax2.grid(True, alpha=0.3)

# Adicionar rótulos aos pontos
for i, curso in enumerate(df_desempenho['Curso']):
    ax2.annotate(curso, (df_desempenho['Nota_Media'][i], df_desempenho['Taxa_Aprovacao'][i]),
                xytext=(5, 5), textcoords='offset points', fontsize=9)

plt.tight_layout()
plt.show()
```

## Resumo Estatístico

```python
# Estatísticas descritivas
print("Resumo Estatístico dos Indicadores:")
print(f"Taxa de Aprovação Média: {df_desempenho['Taxa_Aprovacao'].mean():.1f}%")
print(f"Taxa de Evasão Média: {df_desempenho['Taxa_Evasao'].mean():.1f}%")
print(f"Nota Média Geral: {df_desempenho['Nota_Media'].mean():.1f}")
print(f"\nTotal de Matrículas 2024: {df_cursos['Matriculas_2024'].sum()}")
print(f"Crescimento médio 2022-2024: {((df_cursos['Matriculas_2024'].sum() / df_cursos['Matriculas_2022'].sum()) - 1) * 100:.1f}%")
```
